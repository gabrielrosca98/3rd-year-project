### Load the data and analyse it

```python
import pandas as pd
import tensorflow as tf
import numpy as np
import nltk
import sys
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import DBSCAN, Birch
from sklearn.cluster import OPTICS
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import directed_hausdorff
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
data = pd.read_csv('20190710 - DS Challenge - Breaking News Articles Detection - DSChallengeArticleId.csv')
#Show the first 5 entries of the dataset
#print(data.head())

#Get the number of samples and count the number of missing descriptions for articles
number_samples = data.shape[0]
number_missing_description = data['ArticleDescription'].isna().value_counts()[1]
number_descriptions = number_samples - number_missing_description
print("We have " + str(number_samples) + " articles from which " + str(number_missing_description) + " are missing a description")

plt.figure(figsize=(10, 5))
plt.hist([len(sample) for sample in list(data['ArticleTitle'])], 60)
plt.xlabel('Length of article title')
plt.ylabel('Number of articles')
plt.title('Article title length distribution')
plt.show()

plt.figure(figsize=(10, 5))
plt.hist([len(str(sample)) for sample in list(data['ArticleDescription'])], 150)
plt.xlabel('Length of article title')
plt.ylabel('Number of articles')
plt.title('Article description length distribution')
plt.show()


#Calculate mean description length
total_length = 0
for sample in list(data['ArticleDescription']):
    total_length += len(str(sample))
mean_description_length = total_length / float(number_descriptions)
print("We have a mean description length of: %.2f" %(mean_description_length))

#Calculate mean title lenght
total_length = 0 
for sample in list(data['ArticleTitle']):
    total_length += len(str(sample))
mean_title_length = total_length / float(number_samples)
print("We have a mean title length of: %.2f" %(mean_title_length))

```

### Show the most frequent words(exclude stopwords)

```python
vectorizer = CountVectorizer(stop_words=stopwords.words('english'))
vectorized_texts = vectorizer.fit_transform(list(data['ArticleTitle']))

ngrams = vectorizer.get_feature_names()
ngrams_count = vectorized_texts.sum(axis=0).tolist()[0]

ngrams_df = pd.DataFrame({'ngram':ngrams,'ncount':ngrams_count})
ngrams_df = ngrams_df.sort_values(by='ncount',ascending=False)
number_ngrams = 40
count = list(ngrams_df['ncount'].iloc[:number_ngrams])
ngram = list(ngrams_df['ngram'].iloc[:number_ngrams])
index = np.arange(number_ngrams)
plt.figure(figsize=(35, 20))
plt.title('Ngram frequency distribution throughout the titles(excluded stopwords)')
plt.bar(index, count, width=0.8)
plt.xticks(index, ngram)
plt.xlabel('Ngram')
plt.ylabel('Frequency')
plt.show()
```

### Calculate the TFIDF and identify major events

First we calculate the TFIDF(term frequencyinverse document frequency) in order to obtain the weight of each word in each article in this dataset. Term frequency counts the raw frequency of each word in an article. Inverse document frequency is a measure of how much the word provides.

Fill all the missing descriptions and then merge the two columns: 'ArticleTitle' and 'ArticleDescription'. 

We need to apply an unsupervised data clustering algorithm in order to group and identify major events from the dataset. Our features will be represented by the tfidf vector containing the values for each token(word). Furthermore, we exclude the stopwords("the","to" etc) in order to reduce the computation complexity. As they don't provide much useful information we can safely remove them from the classification.

Use the DBSCAN(Density-based spatial clustering of applications with noise) algorithm to cluster the events. Use the 'cosine' similarity distance and require at least 2 samples in order to create a cluster.

```python
tfidf = TfidfVectorizer(ngram_range=(1,3),dtype='int32',strip_accents='unicode',analyzer='word',min_df=2,stop_words=stopwords.words('english'))
#Fill missing descriptions
data['ArticleDescription'] = data['ArticleDescription'].fillna("")
text = data['ArticleTitle'] + " " + data['ArticleDescription']
tfidf_text = tfidf.fit_transform(text)
clustering = DBSCAN(eps=0.64, min_samples=2, metric='cosine').fit(tfidf_text)
data['EventId'] = clustering.labels_
data.to_csv('20190710 - DS Challenge - Breaking News Articles Detection - DSChallengeArticleId.csv')
```

### Extract information about the events

Use the nltk library to extract the location from the article title. Assume that the first article date is the nearest to the event and use the time/datetime library to convert the UC to a better format.

```python
from nltk import word_tokenize, pos_tag, ne_chunk
from collections import Counter
from datetime import datetime
import time
number_events = data['EventId'].max()
np.set_printoptions(threshold=sys.maxsize)
pd.set_option('display.max_colwidth', -1)
info_events = pd.DataFrame(columns=['EventId', 'EventTitle','EventLocation','EventTime'])
for event_number in range(0, number_events + 1):
    event_info = data[data['EventId'] == event_number]
    #print(event_info['ArticleId'])
    location = {}
    for event_article_number in event_info['ArticleId']:
        chunked =  ne_chunk(pos_tag(word_tokenize(str(event_info[event_info['ArticleId'] == event_article_number]['ArticleDescription']))))
        for chunk in chunked:
            if hasattr (chunk, 'label'):
                if chunk.label() == 'GPE':
                    #print(''.join (c[0] for c in chunk))
                    loc = ''.join (c[0] for c in chunk)
                    location[loc] = location.get(loc, 0) + 1         
    location = sorted(location, key=location.get,reverse=True)
    first_article = time.strftime("%a, %d %b %Y %H:%M:%S", time.localtime(event_info['ArticlePublishedTime'].min()))
    uc_year = 31556926.08
    
    if bool(location):
        info_events = info_events.append({'EventId' : event_number, 'EventTitle' : event_info.iloc[0]['ArticleTitle'], 'EventLocation' : location[0], 'EventTime' : first_article} , ignore_index=True)
    else:
        info_events = info_events.append({'EventId' : event_number, 'EventTitle' : event_info.iloc[0]['ArticleTitle'], 'EventLocation' : "-",  'EventTime' : first_article} , ignore_index=True)
info_events.to_csv("DetailedEventInfo.csv")
```
